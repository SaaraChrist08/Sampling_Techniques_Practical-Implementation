# DATA ANALYSIS

Calculating average satisfaction for each department

We consider a Hypothesis here, 

  H0: Average out of the Total Departments are satisfied with the Placement Cell,
  
  H1: Average out of the Total Departments are not satisfied with the Placement Cell

```python
satisfaction_questions = [
    'How satisfied are you with the overall placement support services provided by Christ Lavasa?',
    'How would you rate the responsiveness of the placement support team in addressing your queries and concerns?',
    'How satisfied are you with the quality and relevance of the job/internship opportunities provided by placement support services?',
    'How satisfied are you with the interview preparation assistance provided by placement support services?'
]

# Defining the weights for each question
weights = [0.3, 0.2, 0.2, 0.1]

df['Composite Satisfaction Score'] = (df[satisfaction_questions] * weights).sum(axis=1)
```

```python
df[['Department', 'Composite Satisfaction Score']]
```

```python
department_satisfaction = df.groupby("Department")['Composite Satisfaction Score'].mean()

department_satisfaction
```

```python
satisfaction_scores_by_department = [df[df['Department'] == department]['Composite Satisfaction Score'] for department in df['Department'].unique()]
```

```python
f_statistic, p_value = f_oneway(*satisfaction_scores_by_department)
print("ANOVA Test:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

The F-statistic of 20.321109 is large, indicating a substantial difference between the mean satisfaction scores of the departments.

The very small p-value ( < 0.001) suggests that the observed differences in satisfaction scores between departments are highly unlikely to be due to random chance alone.

Thus, we reject the Null Hypothesis and accept the Alternative Hypothesis that 'Average out of the Total Departments are not satisfied with the Placement Cell'

Also from the above analysis, we can see that students from the BBA department (1.562687) are the most statisfied and second comes the students from the B.Com Department (1.463158), and the least statisfied are students from School of Law (0.800000), suggesting the Placement Cell is bringing in companies with jobs related to Analytical skills combined with a good Managerial skillset than those in need of man power in the Legal department alone.
This also indicates that the Law graduates are more likely to find their own jobs rather than depending on the Placement Cell as it would make sense for them to start their legal carrer tenure at places of their own choice, so leaving the School of Law aside the second lowest score was of Department of Data Science (1.062745), indicating the companies that are brought by the Placement Cell are less focused on Analytical and Software skill alone.

### Calculating the Overall Average satisfaction of the Students in the Interview Training Programme held by the Placements Cells 

We consider another Hypothesis here,

  H0: All the Departments are equally satisfied with the Interview Training Programmes
  
  H1: All the Departments are not equally satisfied with the Interview Training Programmes

```python
df['Did the placement support services offer any specialized interview preparation resources or materials (e.g., interview guides, practice questions, resume review)?'] = df['Did the placement support services offer any specialized interview preparation resources or materials (e.g., interview guides, practice questions, resume review)?'].map({'Yes': 1, 'No': 0})
```

As we can see there are few non - numeric values in some of the numeric columns so it is important to deal with them and rectify the errors

```python
column_name = 'How many mock interviews or interview preparation sessions did you attend through placement support services before your actual interviews? '
df[column_name] = pd.to_numeric(df[column_name], errors='coerce').fillna(0)
```

```python
df[['Department', 'How many mock interviews or interview preparation sessions did you attend through placement support services before your actual interviews? ']]
```

```python
column_name = 'How many job interviews have you attended as a result of utilizing placement support services in the past year?'
df[column_name] = pd.to_numeric(df[column_name], errors='coerce').fillna(0)
```

```python
df[['How many job interviews have you attended as a result of utilizing placement support services in the past year?']]
```

```python
satisfaction_questions_1 = [
    'How many mock interviews or interview preparation sessions did you attend through placement support services before your actual interviews? ',
    'How satisfied are you with the quality and relevance of the job/internship opportunities provided by placement support services?',
    'How satisfied are you with the interview preparation assistance provided by placement support services?',
    'How many job interviews have you attended as a result of utilizing placement support services in the past year?'
]

# Defining the weights for each question
weights_1 = [0.2, 0.4, 0.3, 0.35]

df['Satisfaction Score_1'] = (df[satisfaction_questions_1] * weights_1).sum(axis=1)
```

```python
department_satisfaction_1 = df.groupby("Department")['Satisfaction Score_1'].mean()

department_satisfaction_1
```

```python
satisfaction_scores_1_by_department = [df[df['Department'] == department]['Satisfaction Score_1'] for department in df['Department'].unique()]
```

```python
f_statistic, p_value = f_oneway(*satisfaction_scores_1_by_department)

print("ANOVA Test:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)
```

The F-statistic, measures the ratio of variation between group means to the variation within the groups. In this case, the F-statistic is 33.947, indicating a substantial difference again

The p-value associated with the F-statistic is very small ( < 0.001). 
This indicates that the observed differences in satisfaction scores between departments are highly unlikely to be due to random chance alone.

Based on these results, we reject the Null Hypothesis and conclude that there is a statistically significant difference in satisfaction scores for the Interview Training programmes across different departments.
